```python
script = """
Hey there and welcome to my YouTube channel! My name is HAMZA, and I'm so excited to have you here with me today.
On this channel, we explore all sorts of interesting and inspiring stories, from real-life experiences to fictional tales that transport us to new worlds.
Today, I want to share a special story with you called 'The Big Deal.' It's a story about perseverance, hard work, and the power of believing in yourself.
So grab a comfy seat, settle in, and let's dive into this amazing tale. Are you ready? Let's go!
As I emptied the fiftieth trash can of the day into the giant black bin, I couldn't help but chuckle. To most people, my job as a janitor in a hospital wasn't glamorous, but to me, it was everything. I took pride in my work, ensuring that each room was spotless and ready for the next patient.
I took off my gloves and threw them in the dumpster on my way out. As I reached for my phone, it rang in my pocket. It was my boss, Jeff, asking if I could stay for another 8 hours. I had already worked a 12-hour shift, but I couldn't let Jeff down. "Fine," I said, "but is it overtime?"
Jeff sighed in relief and said, "Of course, Brian. Always."
Heading back into the hospital, I walked through the halls with ease, dodging carts and frantic nurses. I was used to the chaos by now. I knew each wing of the hospital like the back of my hand.
Entering a new wing, I stepped into a room where a slim girl lay in a hospital bed. She seemed to be asleep, but as I emptied the trash can, she shifted and spoke softly, "Water." I pressed the call light, and a young nurse in pink scrubs appeared before me. "She wants water," I told her.
The nurse was surprised and said, "She hasn't spoken for days." I shrugged and moved on to the next room.
The next room was dimly lit, and I could barely make out the figure in the bed. I was surprised to see that the room had an occupant. I usually got out of dealing with people in every room; I just cleaned and went my way. But this time, I felt drawn to the man in the bed.
His breath rattled deep in his lungs, and I closed my eyes for a moment, feeling the pain of human suffering. The man stirred and his half-closed eyes lit up when he saw me. "Do you know Joni?" he asked.
I nodded, and he smiled, giving me a flash of his yellowed dentures. "Tell her thank you for the brownies."
His eyes fluttered closed, and I waited for the rattle of breath, but it never came. The nurse and doctor that came in rushed me out the same way I had come. This man wouldn't need the room cleaned now, but I'd be back later.
As I walked away, I thought about the last words of that old gentleman. Thanking Joni for brownies—his last words he ever said in his life were a thank you. It struck me that there are all kinds of people in this world, some uppity and too good to look at someone like me, a sanitation worker, cleaner, or even a janitor. But today, I had heard a man's last words, and they were a thank you to another janitor.
As I finished cleaning up a spill in the hallway, an alarm blared over the intercom. I couldn't make out the words over the clamor of doctors rushing past me. It sounded like a trauma patient, but I wasn't sure. A hospital bed with a man...
Thanks for tuning in to my channel and listening to "The Big Deal" inspirational story. I hope it brought some positivity and motivation into your day. If you enjoyed this video, please hit the like button, subscribe to my channel for more content, and don't forget to leave your thoughts and comments below. See you in the next video!
""".replace("\n", " ").strip()
```

```python
import nltk
nltk.download('punkt')
sentences = nltk.sent_tokenize(script)
```

v2/en_speaker_9 - female english
v2/hi_speaker_9 - female hindi

```python
GEN_TEMP = 0.6 #ddefault 0.6
SPEAKER = "v2/en_speaker_9"
silence = np.zeros(int(0))
print(silence)
pieces = []
for sentence in sentences:
    semantic_tokens = generate_text_semantic(
        sentence,
        history_prompt=SPEAKER,
        temp=GEN_TEMP,
        min_eos_p=0.05,  # this controls how likely the generation is to end
    )

    audio_array = semantic_to_waveform(semantic_tokens, history_prompt=SPEAKER,)
    pieces += [audio_array, silence.copy()]
```

```python
from scipy.io.wavfile import write as write_wav

generatedAudio = np.concatenate(pieces)

# save audio to disk
write_wav("bark_generation_female.wav", SAMPLE_RATE, generatedAudio)

#play
Audio(generatedAudio, rate=SAMPLE_RATE)
```

Two person conversation

```python
speaker_lookup = {"Samantha": "v2/en_speaker_9", "John": "v2/en_speaker_2"}

# Script generated by chat GPT
script = """
Samantha: Hey there, John, did you know that fleas can jump 350 times their body length?

John: Wow, that's incredible! Imagine if humans could jump like that.

Samantha: We'd be able to jump over buildings and skip all the traffic.

John: [laughs] Yeah, and we could show off our jumping skills at the Olympics.

Samantha: Speaking of birds, did you know that hummingbirds are the only birds that can fly backwards?

John: No way, really? That's amazing! I wonder how they do that.

Samantha: They have a unique joint in their wings that allows them to rotate them in a full circle.

John: (impressed) That's so cool. I wish I had that kind of flexibility.

Samantha: (jokingly) Maybe you should start doing some yoga.

John: [laughs] I think I'll stick to being a couch potato.

Samantha: (chuckles) Fair enough. Did you know that crocodiles cannot stick their tongue out?

John: Really? Why not?

Samantha: Their tongues are attached to the bottom of their mouth by a membrane, so they can't move it around like we can.

John: (surprised) I had no idea. That's kind of sad though, they must miss out on all the good food flavors.

Samantha: (sympathetically) Yeah, it must be tough. Speaking of brains, did you know that starfish don't have one?

John: (shocked) What?! How do they even function without a brain?

Samantha: They have a complex nervous system that allows them to sense their environment and make decisions.

John: (amazed) That's so interesting. It just goes to show that brains aren't everything.

Samantha: (teasingly) Yeah, and speaking of noses, did you know that slugs have four of them?

John: (disgusted) Ew, that's gross. Why do they need so many noses?

Samantha: They use them to detect different scents and navigate their environment.

John: (sarcastically) Yeah, because smelling things as a slug is so important.

Samantha: [laughs] Hey, you never know when you might need to find your way back to your slimy trail."""
script = script.strip().split("\n")
script = [s.strip() for s in script if s]
script
```

```python
from scipy.io.wavfile import write as write_wav

pieces = []
silence = np.zeros(int(0.5*SAMPLE_RATE))
for line in script:
    speaker, text = line.split(": ")
    audio_array = generate_audio(text, history_prompt=speaker_lookup[speaker], )
    pieces += [audio_array, silence.copy()]
```

```python
generatedAudio = np.concatenate(pieces)

# save audio to disk
write_wav("bark_generation.wav", SAMPLE_RATE, generatedAudio)

#play
Audio(generatedAudio, rate=SAMPLE_RATE)
```
